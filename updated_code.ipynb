```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
print(len(X_train))
print(len(X_test))
X_train.shape, X_test.shape
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.7)
len(set(corr_features))
```

```python
corr_features
```

```python
X_train=X_train.drop(corr_features,axis=1)

```

```python
X_test=X_test.drop(corr_features,axis=1)
```

```python

from sklearn import preprocessing
X_train = preprocessing.normalize(X_train, norm='l2')
X_test = preprocessing.normalize(X_test, norm='l2')
```

```python
X_train
```

```python
X_test
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
clf = tree.DecisionTreeClassifier(min_samples_split=2,min_samples_leaf=20)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

acc_dtree=accuracy_score(pred,y_test)
print(acc_dtree*100)



print(classification_report(y_test,pred))

# sns.heatmap(confusion_matrix(pred,y_test),cmap="Blues",annot=True,fmt='d')
# plt.xlabel("Actual values")
# plt.ylabel("predicted values")
# plt.title("confusion matrix visualization")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=102)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

acc_rforrest=accuracy_score(pred,y_test)
print(acc_rforrest*100)

print(classification_report(y_test,pred))
```

```python
from sklearn import svm
clf = svm.SVC(C=30,kernel="linear",gamma=0.1,max_iter=100,tol= 1.2)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

acc_svm=accuracy_score(pred,y_test)
print(acc_svm*100)

# print(classification_report(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

acc_knn=accuracy_score(pred,y_test)
print(acc_knn*100)

# print(classification_report(y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression() 
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

acc_lr=accuracy_score(pred,y_test)
print(acc_lr*100)


# print(classification_report(y_test,pred))
```
